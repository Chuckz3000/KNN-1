```python

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


df = pd.read_csv('Mall_Customers.csv')



numeric_cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']


scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[numeric_cols])

#  Convert the scaled data back into a DataFrame for easier handling
df_scaled = pd.DataFrame(df_scaled, columns=numeric_cols)

# Step 5: Use K-Means clustering on the scaled data
# Elbow method to find the optimal number of clusters (k)
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    sse.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.show()
```

```python
#  Perform K-Means clustering with the optimal number of clusters
optimal_k = 5
kmeans = KMeans(n_clusters=optimal_k, random_state=42)

# Fit the model on the scaled data
kmeans.fit(df_scaled)

# Add the cluster labels to the original dataframe
df['Cluster'] = kmeans.labels_

# Step 7: Visualize the clusters
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='viridis', s=100)
plt.title(f'Customer Segments (k={optimal_k})')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()

# Step 8: Analyze the clusters
cluster_means = df.groupby('Cluster').mean()

# Display the mean values of each cluster
print(cluster_means)
```

```python
# Step 8: Analyze the clusters
# Exclude non-numeric columns like 'Gender' when calculating cluster means
numeric_cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']  # Columns with numeric data

cluster_means = df.groupby('Cluster')[numeric_cols].mean()

# Display the mean values of each cluster
print(cluster_means)
```

```python
#Task 3: Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
optimal_k = 5
from sklearn.preprocessing import StandardScaler


numeric_features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']


scaler = StandardScaler()


df_scaled = scaler.fit_transform(df[numeric_features])


print(df_scaled[:5])


hierarchical_clustering = AgglomerativeClustering(n_clusters=optimal_k, linkage='ward')

hierarchical_clustering.fit(df_scaled)

# Add the cluster labels to the original dataframe
df['Hierarchical_Cluster'] = hierarchical_clustering.labels_

# Display the first few rows
df.head()
```

```python
#Visualize the Dendrogram that so yo tell me wassup
linked = linkage(df_scaled, method='ward')

plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Hierarchical Clustering')
plt.show()
```

```python
# Task 4:using PCA to reduce the dimensionality
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# Initialize PCA to reduce the data to 2 dimensions
pca = PCA(n_components=2)

# Apply PCA to the scaled data
df_pca = pca.fit_transform(df_scaled)

# Check how much variance is explained by each principal component
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance by each principal component: {explained_variance}")
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=df['Hierarchical_Cluster'], cmap='plasma', s=50)
plt.title('2D PCA Plot with Hierarchical Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
# reducing the dataset to two dimensions with PCA allows for better visualization of how well-defined the clusters are, especially when overlaying them.
```

```python
# Task 5: Dimensionality Reduction with t-SNE.
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
# Initialize t-SNE to reduce the data to 2 dimensions
tsne = TSNE(n_components=2, random_state=42)

# Apply t-SNE to the scaled data
df_tsne = tsne.fit_transform(df_scaled)

# Check the first few rows of the t-SNE transformed data
print(df_tsne[:5])
plt.figure(figsize=(8, 6))
plt.scatter(df_tsne[:, 0], df_tsne[:, 1], c=df['Hierarchical_Cluster'], cmap='plasma', s=50)
plt.title('2D t-SNE Plot with Hierarchical Clusters')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.colorbar(label='Cluster')
plt.show()
# t-SNE is often better at preserving local cluster structures and might show tighter, more separated clusters compared to PCA.
#t-SNE is usually more effective for visualizing non-linear separations.
```
